```python
import numpy as np
import pandas as pd
from linear_regression import LinearRegression
from ransac_regression import RansacRegression
```

```python
data = pd.read_csv('All_Character_Stat.csv')
data.head()
```

```python
data.groupby('Lv').count()
```

```python
data = data[(data['Lv'] != '20+') & (data['Lv'] != '40+') & (data['Lv'] != '50+') & (data['Lv'] != '60+')
            & (data['Lv'] != '80+') & (data['Lv'] != '70+')]
data = data.sample(frac=1)
x = data.drop(['Character', 'Lv'], axis=1).values
y = np.array(data['Lv'].values, dtype='int32')
train_x, train_y = x[:(4 * x.shape[0] // 5)], y[:(4 * x.shape[0] // 5)]
test_x, test_y = x[(4 * x.shape[0] // 5):], y[(4 * x.shape[0] // 5):]
model = LinearRegression()
ransac_model = RansacRegression()
model.fit(train_x, train_y)
print("Точность на тренировочных данных:", model.score(train_x, train_y))
print("Точность на тестовых данных:", model.score(test_x, test_y))
model.fit(train_x, train_y, l2_regulation=1)
print("Точность на тестовых данных с L2 регуляризацией", model.score(test_x, test_y))
print("Тончость на cross валидации:", model.cross_val_score(x, y, cv=10))
ransac_model.fit(train_x, train_y, 2)
print("Тончость ransac на тренировочных данных:", ransac_model.score(train_x, train_y))
print("Тончость ransac на тестовых данных:", ransac_model.score(test_x, test_y))
```
