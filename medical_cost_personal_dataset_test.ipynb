```python
import numpy as np
import pandas as pd
from LinearRegression import LinearRegression
from LinearRegression import RansacRegression
```

```python
data = pd.read_csv('insurance.csv')
data.head()
```

```python
data.loc[data['smoker'] == 'yes', 'smoker'] = 1
data.loc[data['smoker'] == 'no', 'smoker'] = 0
data.loc[data['sex'] == 'male', 'sex'] = 1
data.loc[data['sex'] == 'female', 'sex'] = 0
```

```python
data.groupby('region').count()
```

```python
data['north'] = ((data['region'] == 'northeast') | (data['region'] == ('northwest')))
data['west'] = ((data['region'] == 'southwest') | (data['region'] == ('northwest')))
data = data.drop(['region'], axis=1)
data.loc[data['north'], 'north'] = 1
data.loc[data['north'] == False, 'north'] = 0
data.loc[data['west'], 'west'] = 1
data.loc[data['west'] == False, 'west'] = 0
```

```python
data['sex'] = data['sex'].astype('int64')
data['smoker'] = data['smoker'].astype('int64')
data['west'] = data['west'].astype('int64')
data['north'] = data['north'].astype('int64')
```

```python
data.head()
```

```python
data = data.sample(frac=1)
x = data.drop(['charges'], axis=1).values
y = np.array(data['charges'].values, dtype='int32')
train_x, train_y = x[:(4 * x.shape[0] // 5)], y[:(4 * x.shape[0] // 5)]
test_x, test_y = x[(4 * x.shape[0] // 5):], y[(4 * x.shape[0] // 5):]
model = LinearRegression()
ransac_model = RansacRegression()
model.fit(train_x, train_y)
print("Точность на тренировочных данных:", model.score(train_x, train_y))
print("Точность на тестовых данных:", model.score(test_x, test_y))
model.fit(train_x, train_y, l2_regulation=1)
print("Точность на тестовых данных с L2 регуляризацией", model.score(test_x, test_y))
print("Тончость на cross валидации:", model.cross_val_score(x, y, cv=10))
ransac_model.fit(train_x, train_y, 600)
print("Тончость ransac на тренировочных данных:", ransac_model.score(train_x, train_y))
print("Тончость ransac на тестовых данных:", ransac_model.score(test_x, test_y))
```
